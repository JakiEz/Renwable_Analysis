```python
import pandas as pd
import numpy as np

df = pd.read_csv('../datasets/04 share-electricity-renewables.csv')
```

```python
df['Year'] = df['Year'].astype(int)

df_filtered = df[(df['Year'] >= 1985) & (df['Year'] <= 2021)]

df_filtered = df_filtered[df_filtered['Entity'] != 'World']

df_filtered = df_filtered[df_filtered['Code'].notna() & (df_filtered['Code'].str.len() == 3)]

country_year_counts = df_filtered.groupby('Entity')['Year'].nunique()

complete_countries = country_year_counts[country_year_counts == 37].index.tolist()

df_clean = df_filtered[df_filtered['Entity'].isin(complete_countries)]

df_clean = df_clean.sort_values(['Entity', 'Year'])

df_clean.to_csv('clean_countries_renewables_1985_2021.csv', index=False)

print(f"Original dataset size: {len(df)} rows")
print(f"Cleaned dataset size: {len(df_clean)} rows")
print(f"Number of countries with complete data: {len(complete_countries)}")
print("Countries with complete data:")
for country in sorted(complete_countries):
    print(f"- {country}")
```

```python
# Count how many years each entity has data for
entity_year_counts = df_filtered.groupby('Entity')['Year'].nunique()

# Find entities that have data for all 37 years (1985-2021)
complete_entities = entity_year_counts[entity_year_counts == 37].index.tolist()

# Print the number of entities with complete data
print(f"Number of entities with complete data: {len(complete_entities)}")
```
